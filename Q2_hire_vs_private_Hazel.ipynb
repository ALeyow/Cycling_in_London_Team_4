# Aggregate by Year - Inner & Central

```python
# Aggregate table
london_ci_year = london_clean.groupby(['Year']).sum().reset_index().copy()
```

```python
def percentage(col1,col2):
    return (col1 / col2) * 100
```

```python
# Calculate percentage of private & Public cycle hires out of total cycles
london_ci_year['Private Cycles %'] = percentage(london_ci_year['Number of private cycles'],london_ci_year['Total cycles']).round(0)  
london_ci_year['Public Cycles hires %'] =  percentage(london_ci_year['Number of cycle hire bikes'],london_ci_year['Total cycles']).round(0)   
```

```python
# Print table
london_ci_year
```

```python
sns.set(rc = {'figure.figsize':(25,10)}, font_scale = 10)
sns.set_theme(style="ticks")
y = sns.lineplot(data = london_ci_year, x = 'Year', y = 'Total cycles', color = 'red', ci = None)
y.set_title('Total Cycles per Year in Central & Inner London', fontsize = 30)
y.set_xlabel('Survey Year', fontsize = 20)
y.set_ylabel('Total Cycles (M)', fontsize = 20)


plt.savefig('total cycles per year in central inner london.png', 
           transparent=True)
```

```python
# Difference when using full dataset?
sns.set(rc = {'figure.figsize':(25,10)}, font_scale = 10)
sns.set_theme(style="ticks")
y = sns.lineplot(data = london_clean, x = 'Year', y = 'Total cycles', color = 'red', ci = None)
y.set_title('Total Cycles per Year in Central & Inner London', fontsize = 30)
y.set_xlabel('Survey Year', fontsize = 20)
y.set_ylabel('Total Cycles (M)', fontsize = 20)
```

```python
sns.set(rc = {'figure.figsize':(25,10)}, font_scale = 10)
sns.set_theme(style="ticks")
y = sns.lineplot(data = london_ci_year, x = 'Year', y = 'Number of cycle hire bikes', color = 'Green', ci = None)
y.set_title('Cycles hires per Year in Central & Inner London', fontsize = 30)
y.set_xlabel('Survey Year', fontsize = 20)
y.set_ylabel('# of Cycles hires', fontsize = 20)

plt.savefig('cycles hires per year in central inner london.png', 
           transparent=True)
```

Overall, total cycles increased each year from 2014 to 2019, then there was a sharp decrease in 2020 which may be due to the pandemic lockdown restrictions.

Private cycles acccounted for over 90% of the total cycles, whereas cycle hires accounted for 10% of total cycles only. This means majority of those who chooses to cycle uses their own bicycles. To increase the uptake of cycling, the government should target citizens using other modes of transport and convert them to cycling to increase cycle hires.

In central london, more people hire cycles when travelling towards eastbound and westbound, while more people use private cycles when travelling towards north and south bound.
In inner london, most people hire cycles or use private cycles when travlling towards in northbound and southbound. This means people are less likely to cycle when travelling towards east and west bound in inner london.

Is this due to the lack of infrastructure in in central london towards north and southbound?
